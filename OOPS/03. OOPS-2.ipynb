# Polymorphism

```python
# Parent Class
class Vehicle:
  current_year = 2021
  base_price = 1000

  def __init__(self,make,model,fuel):
    self.make = make
    #private
    self.__model = model
    self.__fuel = fuel
  
  def get_value(self):
    age = Vehicle.current_year - self.__model
    print("Parent Method")
    return Vehicle.base_price*(1/age)
```

```python
class Car(Vehicle):
  def __init__(self,make,model,fuel,air_conditioning,sunroof):
    super(Car,self).__init__(make,model,fuel)
    
    # Child Specific Attribute
    self.air_conditioning = air_conditioning
    self.sunroof = sunroof

  def get_value(self):
    Vehicle.base_price = 5000
    age = Vehicle.current_year - self._Vehicle__model
    print("Child Method")
    return Vehicle.base_price*(1/age)
```

```python
obj = Car('Tesla',2019,"Electric",True,True)
```

```python
obj.get_value()
```

```python
parent_obj = Vehicle("Ford",2019,"Petron")
```

```python
parent_obj.get_value()
```

```python
class Vehicle:

  # Class Attribute
  current_year = 2021
  base_price = 1000

  def __init__(self,make,model,fuel):
    self.make = make
    self.__model = model
    self.fuel = fuel

  def get_value(self):
    age = Vehicle.current_year - self.__model
    print("Default Parent Method")
    return Vehicle.base_price * 1/age
```

```python
class Car(Vehicle):
  def __init__(self,make,model,fuel,air_conditioning,sunroof):
    super(Car,self).__init__(make,model,fuel)
    self.air_conditioning = air_conditioning
    self.sunroof = sunroof

  def get_value(self):
    Vehicle.base_price = 5000
    age = Vehicle.current_year - self._Vehicle__model
    print("This is overriddden")
    return Vehicle.base_price*(1/age)
```

```python
obj = Car(1,2,3,4,5)
```

```python
obj.get_value()
```

# Protected Member

just a single underscore

```python
class Vehicle:

  # Class Attribute
  current_year = 2021
  base_price = 1000

  def __init__(self,make,model,fuel):
    self.make = make

    #Protectec
    self._model = model
    self._fuel = fuel

  # Protected Method
  def _get_value(self):
    age = Vehicle.current_year - self._model
    print("Default Parent Method")
    return Vehicle.base_price * 1/age
```

```python
class Car(Vehicle):
  def __init__(self,make,model,fuel,air_conditioning,sunroof):
    super(Car,self).__init__(make,model,fuel)

    #Protected
    self._air_conditioning = air_conditioning
    self._sunroof = sunroof

  # Overridig Method
  def _get_value(self):
    Vehicle.base_price = 5000
    age = Vehicle.current_year - self._model
    print("This is overriddden")
    return Vehicle.base_price*(1/age)
```

```python
obj = Car("Tesla",2019,"electric",True,True)
```

```python
obj.__dict__
```

```python
obj._sunroof
```

```python
obj._get_value()
```

# Object Class

All class inherit from object classes. By default there are 3 methods that come along with object class. They are :


1.   __init__
2.   __new__
3. __str__


```python
class Circle(object):
  def __init__(self,rad):
    self.rad = rad

  def __str__(self):
    return "CIRCLE CLASS INITIALIZED"
```

```python
c= Circle(10)
print(c)
```

# Multilevel Inheritence

```python
class Vehicle:

  # Class Attribute
  current_year = 2021
  base_price = 1000

  def __init__(self,make,model,fuel):
    self.make = make
    self.model = model
    self._fuel = fuel

  # Protected Method
  def _get_value(self):
    age = Vehicle.current_year - self.model
    print("Default Parent Method")
    return Vehicle.base_price * 1/age
```

```python
class Car(Vehicle):
  def __init__(self,make,model,fuel,air_conditioning,sunroof):
    super(Car,self).__init__(make,model,fuel)
    self.air_conditioning = air_conditioning
    self._sunroof = sunroof

  # Overridig Method
  def _get_value(self):
    Vehicle.base_price = 5000
    age = Vehicle.current_year - self.model
    print("This is overriddden")
    return Vehicle.base_price*(1/age)
```

```python
class ElectricVehicle(Car):
  def __init__(self,make,model,fuel,air_conditioning,sunroof,distance):
    super(ElectricVehicle,self).__init__(make,model,fuel,air_conditioning,sunroof)

    self.distance = distance
```

```python
my_obj = ElectricVehicle("tesla",2019,"Electric",True,True,500)
```

```python
my_obj.__dict__
```

# Method Resolution Order

```python
class Car:
  def __init__(self,make,model,fuel):
    self.make=make
    self.model=model
    self.fuel=fuel

  def get_car_detail(self):
    return "THE MAKE OF THE CAR IS ",self.make," FROM CAR CLASS"

class ElectricCar:
  def __init__(self,make,model,fuel):
    self.make=make
    self.model=model
    self.fuel=fuel

  def get_car_detail(self):
    return "THE MAKE OF THE CAR IS ",self.make," FROM ELECTRIC CAR CLASS"
```

```python
class Taxi(Car,ElectricCar):             #Multiple Inheritence
  def __init__(self,make,model,fuel):
    super().__init__(make,model,fuel)                 
```

```python
obj = Taxi("tesla",2019,True)
```

```python
obj.get_car_detail()
```

```python
Taxi.mro()
```

# OPERATOR OVERLOADING

```python
class Vehicle:
  def __init__(self,make,model,fare):
    self.make = make
    self.model = model
    self.fare = fare

  def __str__(self):
    return "The make is {} and model is {} and fare is {}".format(self.make,self.model,self.fare)

  def __add__(self,other):
    return self.fare+other.fare
```

```python
obj1 = Vehicle('tesla',2019,49)
obj2 = Vehicle('ford',2018,40)
```

```python
print(obj1)
print(obj2)
```

```python
# See Operator Overloading
obj1 + obj2
```
