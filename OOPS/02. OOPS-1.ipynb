# Class Method


1.   Class Metá¸¥od
2.   Static Method
3.   Instance Method



```python
class ExpenseTracker:

  #Class Attribute                :   global
  expense_tracker_version = 0.1

  def __init__(self,tracker_category,opening_balance,budget):
    #Instance/ Object Attribute                 :  specific to the instance
    self.tracker_category = tracker_category
    self.opening_balance = opening_balance
    self.budget = budget

  # Instance Method 1               :      points to the instance  (can access attributes)
  def get_opening_balance(self):
    return self.opening_balance

  # Instance Method 2
  def check_balance(self,limit=100):
    if self.budget>=limit:
      return True
    else:
      return 'Your opening balance is less than limit.'

  # Static Method                   :     doesnt receives any self reference. It is free to take any number of parameter. Since no self i cant access instance related attribute
  @staticmethod
  def convert_amount(amt):
    return float(amt)

  #Class Method or Factory Method
  @classmethod
  def get_attributes_fromstring(cls,diary_entry:str):
    tracker_category,opening_balance,budget = diary_entry.split()
    
    return ExpenseTracker(tracker_category.capitalize(),
                          cls.convert_amount(opening_balance),
                          cls.convert_amount(budget))
```

```python
home_tracker = ExpenseTracker(tracker_category='home',opening_balance=1000,budget=999)

home_tracker.get_opening_balance()
```

```python
home_tracker.check_balance(9999)
```

```python
home_tracker.convert_amount(100)
```

# Access Modifiers


1.   Public Modifier
2.   Private Modifier



```python
class ExpenseTracker:

  #Class Attribute                :   global
  expense_tracker_version = 0.1

  def __init__(self,tracker_category,opening_balance,budget):
    #Instance/ Object Attribute                 :  specific to the instance

    #Public Attribute
    self.tracker_category = tracker_category

    #Private Attribute
    self.__opening_balance = opening_balance
    self.__budget = budget

  # Instance Method 1               :      points to the instance  (can access attributes)
  #Private Method
  def __get_opening_balance(self):
    return self.__opening_balance

  # Instance Method 2
  def check_balance(self,limit=100):
    if self.budget>=limit:
      return True
    else:
      return 'Your opening balance is less than limit.'

  # Static Method                   :     doesnt receives any self reference. It is free to take any number of parameter. Since no self i cant access instance related attribute
  @staticmethod
  def convert_amount(amt):
    return float(amt)

  #Class Method or Factory Method
  @classmethod
  def get_attributes_fromstring(cls,diary_entry:str):
    tracker_category,opening_balance,budget = diary_entry.split()
    
    return ExpenseTracker(tracker_category.capitalize(),
                          cls.convert_amount(opening_balance),
                          cls.convert_amount(budget))
```
