# Abstract Class

```python
from abc import ABC,abstractmethod
```

```python
class Vehicle(ABC):
  @abstractmethod
  def get_value(self):
    pass


class Car(Vehicle):
  def __init__(self,make,model,fare):
    self.make = make
    self.model = model
    self.fare = fare
```

```python
# Didnt use abstract method in derived class
obj = Car("Tesla",2019,40)
```

##### Correct it

```python
class Vehicle(ABC):
  @abstractmethod
  def get_value(self):
    pass


class Car(Vehicle):
  def __init__(self,make,model,fare):
    self.make = make
    self.model = model
    self.fare = fare

  def get_value(self):
    return 1000*self.fare
```

```python
obj = Car("Tesla",2019,40)
obj.get_value()
```

```python
obj = Vehicle()
```

###### Some More Examples

```python
from abc import ABC,abstractmethod
```

```python
class TextReaderAbstract(ABC):
  def __init__(self,path,filename):
    self.path = path
    self.filename = filename

  @abstractmethod
  def get_path(self):
    pass
  
  @abstractmethod
  def get_filename(self):
    pass
```

```python
class TextReader(TextReaderAbstract):

  def get_path(self):
    return self.path
  
  def get_filename(self):
    return self.filename
```

```python
obj = TextReader('user/download','hello.png')
```

```python
print("PATH : ",obj.get_path())
print("FILENAME : ",obj.get_filename())
```

###### Lets do something more

```python
class Bank(ABC):
  @abstractmethod
  def get_interest(self):
    pass
```

```python
class HDFC(Bank):
  def get_interest(self):
    return 8.9
```

```python
obj = HDFC()
obj.get_interest()
```

###### Little More

```python
from abc import ABC,abstractmethod

class A(ABC):

    @abstractmethod
    def fun1(self):
        print('function of class A called')

    @abstractmethod
    def fun2(self):
        pass

class B(A):
    def fun1(self):
        print('function 1 called')
    def fun2(self):
        print('function 2 called')
o = B()
o.fun1()
```

#### A last 1

```python
from abc import ABC,abstractmethod

class A(ABC):

    @abstractmethod
    def fun1(self):
        print('function of class A called')

    @abstractmethod
    def fun2(self):
        pass

class B(A):
    def fun1(self):
        super().fun1()
    def fun2(self):
        print('function 2 called')
o = B()
o.fun1()
```
